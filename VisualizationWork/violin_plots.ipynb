## Introduction

### What is a violin plot? 

Violin plots are distribution plots that also show a five number summary. 

It's like a box plot and density plot combined. The wider the section of the plot, the higher probability it is for population members to be a given value. 

<img src="https://images.ctfassets.net/fi0zmnwlsnja/sdfgtcRp16wTNOcRceGQm/5bfcb73d2261d49ff20dd7857e0152b1/Screen_Shot_2019-03-01_at_11.36.10_AM.png" height="auto" width="300">

The white dot represents the median, while the thick bar shows the interquartile range. 

<img src="https://miro.medium.com/max/650/1*TTMOaNG1o4PgQd-e8LurMg.png" height="auto" width="400">

They are mainly used for: 
   1. Visualizing distribution and shape of data
   2. Easily comparing many different distributions
   

# Violin Plots in Seaborn

To create a violin plot with python, we can use Seaborn and its .violinplot() function. To get started, first import seaborn and load in a dataset:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# We will be using the titanic.csv file from seaborn library
titanic_dataset = sns.load_dataset('titanic')
print(titanic_dataset.head(5))
```

```python
# Example 1

df = sns.load_dataset('titanic')

sns.violinplot( x = 'sex',
            y = 'age',
            data = df,
            color = "Yellow")
 
# Show the plot
plt.show()
```

We can also very easily orient this plot horizontally in seaborn by just switching x and y.

```python
sns.violinplot( x = 'age',
            y = 'sex',
            data = df,
            color = "Yellow")
 
plt.show()
```

# Comparing violin plots

It is very common to want to compare distribution of related categories (for example comparing by gender or age). For this, we can set the "hue" arguement to our differentiating category. 

Here we are visualizing the age distribution of men vs women among each of the classes, using a grouped violin plot. 

```python
# Example 2

df = sns.load_dataset('titanic')
violin = sns.violinplot(
            x = 'class',
            y = 'age',
            hue = 'sex',
            data = df,
            palette = "Blues")
violin.set_title('Distribution of ages for different classes', fontsize=16);

plt.legend(loc='upper right')
plt.show()
```

We can also set **split=True** to get a more compact view of the comparison. This is called a split violin plot.

```python
violin = sns.violinplot(
            x = 'class',
            y = 'age',
            hue = 'sex',
            data = df,
            palette = "Blues",
            split = True)
plt.show()
```

#### What are some alternative visualizations?
There are many alternate plots for viewing distibution, for example density plots, histograms, and box plots. 

Histograms allow us to see how data may fall according to different intervals and are good at showing frequency distribution. 

However they, along with density plots, lack quick visualization of how distributions line up against a five number summary. On the other hand, box plots show this, but lack distribution visualization.

Ridgeline plots may be one of the more useful alternatives if you want to compare multiple distributions fast.
